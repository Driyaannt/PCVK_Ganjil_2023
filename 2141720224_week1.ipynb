```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
img_4 = cv.imread('./images/KTP.jpeg')
img_5 = cv.cvtColor(img_4, cv.COLOR_BGR2RGB)
cv2_imshow(img_4)
plt.imshow(img_5)
```

```python
img_4s = np.copy(img_4)
#tutup bagian nama
for y in range (190,400):
  for x in range (190,280):
    img_4s[x,y] = [255,255,255]

#tutup bagian NIK
img_4ss = np.copy(img_4s)
for y in range (190,500):
  for x in range (320,345):
    img_4ss[x,y] = [255,255,255]
final_frame = cv.hconcat((img_4, img_4ss))
cv2_imshow(final_frame)
```
